Download caltech-101 dataset

```python
import kagglehub
import shutil
import os

# Download dataset (by default into kagglehub cache directory)
path = kagglehub.dataset_download("imbikramsaha/caltech-101")

# Get current working directory (where the notebook is located)
current_dir = os.getcwd()

# Define target directory inside the current notebook folder
target_dir = os.path.join(current_dir, "caltech-101")

# Move dataset from kagglehub cache to the notebook directory
shutil.move(os.path.join(path, "caltech-101"), target_dir)

# Remove existing target directory if it already exists
if os.path.exists(path):
    shutil.rmtree(path)

print("Dataset moved to:", target_dir)
```

Download Brain Tumor dataset

```python
import kagglehub
import os
import shutil

# Download latest version
path = kagglehub.dataset_download("masoudnickparvar/brain-tumor-mri-dataset")

print("Path to dataset files:", path)

current_dir = os.getcwd()

target_dir = os.path.join(current_dir, "brain-tumor")

# Move dataset from kagglehub cache to the notebook directory
for folder_name in os.listdir(path):
    # print(folder_name)
    folder_path = os.path.join(target_dir, folder_name)
    #os.makedirs(folder_path, exist_ok=True)
    shutil.move(os.path.join(path, folder_name), folder_path)

# Remove existing target directory if it already exists
if os.path.exists(path):
    shutil.rmtree(path)
```
